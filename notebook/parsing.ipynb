# PDF Document Parsing

Objective: To accurately extract text with its orginal structure from the provided PDF documents using LlamaParse

Output Format: Markdown

Remarks: Due to the limited free credits of LlamaParse, each document is processed only once using this Notebook and stored in a designated folder for future use.

```python
import os
import sys

parent_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
if parent_dir not in sys.path:
    sys.path.append(parent_dir)

from src.module.parsing import PDFLlamaParser
from src.module.constants import (
    DATA_DIR as data,
    RAW_DOC_DIR as raw, 
    PARSED_DOC_DIR as parsed
)
```

Specify the following directories:
- DATA_DIR: The directory contain RAW_DOC_DIR and PARSED_DOC_DIR
- RAW_DOC_DIR: The parent directory that contains the source PDF documents.
- PARSED_DOC_DIR: The parent directory designated for storing the parsed output files.

```python
DATA_DIR = os.path.join(parent_dir, data)
RAW_DOC_DIR = os.path.join(DATA_DIR, raw)
PARSED_DOC_DIR = os.path.join(DATA_DIR, parsed)

LLAMA_PARSE_TIER = 'agentic_plus'
LLAMA_PARSE_VERSION = "2026-03-02"
LLAMA_PARSE_PROMPT = """
### Role
You are an expert in reading energy industry tender documents, including technical specifications and bid documents.

### Task
Your task is to convert complex PDFs into high-fidelity, semantically structured Markdown with the following requirements

### Requirements
1. Preserve the document hierarchy by maintaining the exact nesting of headers (#, ##, ###). 
2. Ensure every clause and sub-clause (e.g. 4.2.1.a) is correctly indented and structured.
3. Maintain table integrity:
   - Do not simplify tables. Preserve every column and row exactly.
   - Extract ALL tables using GitHub Flavored Markdown (GFM). 
   - If a table spans multiple pages, merge them into one continuous Markdown table. Do not repeat headers mid-table and treat it as one logical entity.
   - If a cell contains a sub-table or complex nested data, encapsulate that entire sub-structure within the single corresponding parent cell using HTML-style line breaks (<br>).
4. Ignore images and diagrams. Focus only on text and tabular data.
5. Ensure technical units (e.g. MW, kV, m/s) and price are parsed correctly without character corruption.
6. Strictly ignore all left, right and center-aligned headers and footers, including page number and document name at the footer
"""
```

```python
# Init PDFLlamaParser
pdf_parser = PDFLlamaParser(LLAMA_PARSE_TIER, LLAMA_PARSE_VERSION, LLAMA_PARSE_PROMPT)

# Parse documents
await pdf_parser.parse_all_pdfs(RAW_DOC_DIR, PARSED_DOC_DIR, skip_dirs=['Drawings'])
```
